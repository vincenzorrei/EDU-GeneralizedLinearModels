```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Dati sintetici
np.random.seed(0)
X = np.random.rand(50, 1) * 10
y = 2.5 * X.flatten() + np.random.randn(50) * 5

# Modello
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Grafico
plt.figure(figsize=(8, 5))
sns.scatterplot(x=X.flatten(), y=y, label="Dati osservati", color='dodgerblue')
plt.plot(X, y_pred, color='red', label="Linea di regressione")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Relazione Lineare tra X e Y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# Dati sintetici
np.random.seed(0)
X = np.linspace(-10, 10, 100).reshape(-1, 1)
y_prob = 1 / (1 + np.exp(-X))

# Grafico
plt.figure(figsize=(8, 5))
plt.plot(X, y_prob, color="dodgerblue", label="Funzione logit (sigmoide)")
plt.axhline(0.5, color='red', linestyle='--', label='Soglia decisionale')
plt.xlabel("X")
plt.ylabel("P(Y=1|X)")
plt.title("Probabilit predetta da una regressione logistica")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Dati sintetici binari
np.random.seed(0)
X = np.linspace(0, 10, 100)
y_binary = (X > 5).astype(int)  # soglia arbitraria

# Regressione lineare spuriously applicata
y_pred = 0.1 * X + 0.2  # solo a scopo illustrativo

plt.figure(figsize=(8, 5))
plt.scatter(X, y_binary, color='dodgerblue', label='Y binaria (0 o 1)', alpha=0.6)
plt.plot(X, y_pred, color='red', label='Previsione lineare')
plt.axhline(1, linestyle='--', color='gray', alpha=0.5)
plt.axhline(0, linestyle='--', color='gray', alpha=0.5)
plt.title("Errore della regressione lineare su variabile binaria")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, poisson

x = np.linspace(-3, 3, 100)

# Normal
normal_pdf = norm.pdf(x, loc=0, scale=1)

# Bernoulli (p = sigmoid(x))
sigmoid = 1 / (1 + np.exp(-x))
bernoulli_pmf = sigmoid  # probabilit di successo

# Poisson ( = e^x per rendere positiva)
poisson_x = np.arange(0, 10)
poisson_lambda = np.exp(1.2)  # esempio arbitrario
poisson_pmf = poisson.pmf(poisson_x, mu=poisson_lambda)

# Plot
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Normale
axs[0].plot(x, normal_pdf, label='Normale', color='blue')
axs[0].set_title("Output Gaussiano")
axs[0].grid(True)

# Bernoulli
axs[1].plot(x, bernoulli_pmf, label='Bernoulli (sigmoide)', color='green')
axs[1].set_title("Output Bernoulliano")
axs[1].set_ylim([0, 1])
axs[1].grid(True)

# Poisson
axs[2].bar(poisson_x, poisson_pmf, color='purple')
axs[2].set_title("Output Poissoniano (conteggi)")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, poisson, gamma as gamma_dist, binom

# Consistent style settings
line_color = 'dodgerblue'
figsize = (8, 5)
grid_kwargs = dict(linestyle='--', alpha=0.7)

np.random.seed(1)

# 1. Normal (Identity link) - assume  = X for demo
X = np.linspace(-5, 5, 200)
mu = X  # identity
plt.figure(figsize=figsize)
plt.plot(X, mu, color=line_color, lw=2, label=" = X")
X_obs = np.random.uniform(-5, 5, 120)
y_obs = X_obs + np.random.normal(0, 1, size=X_obs.size)
plt.scatter(X_obs, y_obs, alpha=0.5, s=20, label="osservazioni simulate")
plt.title("Normal (Identity link)")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True, **grid_kwargs)
plt.legend()
plt.tight_layout()
plt.show()

# 2. Bernoulli (Logit link)
X = np.linspace(-6, 6, 300)
p = 1 / (1 + np.exp(-X))
plt.figure(figsize=figsize)
plt.plot(X, p, color=line_color, lw=2, label="p = (X)")
X_obs = np.random.uniform(-6, 6, 200)
p_obs = 1 / (1 + np.exp(-X_obs))
y_obs = bernoulli.rvs(p_obs)
plt.scatter(X_obs, y_obs,
            alpha=0.5, s=20, label="osservazioni simulate")
plt.title("Bernoulli (Logit link)")
plt.xlabel("X")
plt.ylabel("y (0/1)")
plt.ylim(-0.1, 1.1)
plt.grid(True, **grid_kwargs)
plt.legend()
plt.tight_layout()
plt.show()

# 3. Poisson (Log link)
X = np.linspace(-2, 2, 200)
mu = np.exp(X)
plt.figure(figsize=figsize)
plt.plot(X, mu, color=line_color, lw=2, label=" = e^X")
X_obs = np.random.uniform(-2, 2, 200)
mu_obs = np.exp(X_obs)
y_obs = poisson.rvs(mu_obs)
plt.scatter(X_obs, y_obs, alpha=0.5, s=20, label="osservazioni simulate")
plt.title("Poisson (Log link)")
plt.xlabel("X")
plt.ylabel("conteggi y")
plt.grid(True, **grid_kwargs)
plt.legend()
plt.tight_layout()
plt.show()

# 4. Gamma (Log link)
X = np.linspace(-1, 2.5, 200)
mu = np.exp(X)
plt.figure(figsize=figsize)
plt.plot(X, mu, color=line_color, lw=2, label=" = e^X")
X_obs = np.random.uniform(-1, 2.5, 200)
mu_obs = np.exp(X_obs)
shape_k = 2.0
scale_obs = mu_obs / shape_k
y_obs = gamma_dist.rvs(a=shape_k, scale=scale_obs)
plt.scatter(X_obs, y_obs, alpha=0.5, s=20, label="osservazioni simulate")
plt.title("Gamma (Log link)")
plt.xlabel("X")
plt.ylabel("y positivo")
plt.grid(True, **grid_kwargs)
plt.legend()
plt.tight_layout()
plt.show()

# 5. Binomiale (Logit link, n=10)
X = np.linspace(-6, 6, 300)
p = 1 / (1 + np.exp(-X))
plt.figure(figsize=figsize)
plt.plot(X, p, color=line_color, lw=2, label="p = (X)")
n_trials = 10
X_obs = np.random.uniform(-6, 6, 200)
p_obs = 1 / (1 + np.exp(-X_obs))
successes = binom.rvs(n_trials, p_obs)
prop_obs = successes / n_trials
plt.scatter(X_obs, prop_obs + np.random.uniform(-0.02, 0.02, size=prop_obs.size),
            alpha=0.5, s=20, label="osservazioni simulate (successi/n)")
plt.title("Binomiale (Logit link, n=10)")
plt.xlabel("X")
plt.ylabel("proporzione di successi")
plt.ylim(-0.1, 1.1)
plt.grid(True, **grid_kwargs)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, poisson, gamma, binom

# NORMAL  PDF
x_vals = np.linspace(-4, 4, 1000)
plt.figure(figsize=(8, 5))
plt.plot(x_vals, norm.pdf(x_vals), color='dodgerblue')
plt.title("Distribuzione Normale (=0, =1)")
plt.xlabel("Valori")
plt.ylabel("Densit di probabilit")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Set random seed for reproducibility
np.random.seed(42)

# NORMAL  link identity
X = np.linspace(-4, 4, 100)
beta0, beta1 = 0, 1
mu = beta0 + beta1 * X
Y = mu + np.random.normal(scale=1, size=X.size)

plt.figure(figsize=(8, 5))
plt.scatter(X, Y, label="osservazioni simulate", alpha=0.7)
plt.plot(X, mu, label=" =  = X", linewidth=2, color='dodgerblue')
plt.title("Normale  link identit")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# BERNOULLI PMF
p_fixed = 0.7
plt.figure(figsize=(8, 5))
plt.bar([0, 1], bernoulli.pmf([0, 1], p_fixed), color='dodgerblue')
plt.xticks([0, 1])
plt.title("Distribuzione Bernoulli (p = 0.7)")
plt.xlabel("Esito")
plt.ylabel("Probabilit")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

# Bernoulli  link logit
Xb = np.linspace(-6, 6, 400)
beta0_b, beta1_b = -1, 0.8
eta_b = beta0_b + beta1_b * Xb
p = 1 / (1 + np.exp(-eta_b))
Yb = np.random.binomial(1, p)

plt.figure(figsize=(8, 5))
plt.scatter(Xb, Yb, alpha=0.4, label="osservazioni simulate")
plt.plot(Xb, p, linewidth=2, label="p(X) = logistic(X)", color='dodgerblue')
plt.title("Bernoulli  link logit")
plt.xlabel("X")
plt.ylabel("Probabilit di successo")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# POISSON PMF
k = np.arange(0, 15)
lam = 3
plt.figure(figsize=(8, 5))
plt.bar(k, poisson.pmf(k, lam), color='dodgerblue')
plt.title("Distribuzione Poisson ( = 3)")
plt.xlabel("Conteggi")
plt.ylabel("Probabilit")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

# Poisson  link log
Xp = np.linspace(-2, 2, 400)
beta0_p, beta1_p = 1, 1.2
mu_p = np.exp(beta0_p + beta1_p * Xp)
Yp = np.random.poisson(mu_p)

plt.figure(figsize=(8, 5))
plt.scatter(Xp, Yp, alpha=0.4, label="osservazioni simulate")
plt.plot(Xp, mu_p, linewidth=2, label=" = exp(X)", color='dodgerblue')
plt.title("Poisson  link log")
plt.xlabel("X")
plt.ylabel("Conteggi Y")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# GAMMA PDF
xg = np.linspace(0.001, 20, 1000)
k_shape, theta_scale = 2, 2
plt.figure(figsize=(8, 5))
plt.plot(xg, gamma.pdf(xg, a=k_shape, scale=theta_scale), color='dodgerblue')
plt.title("Distribuzione Gamma (shape = 2, scale = 2)")
plt.xlabel("Valori")
plt.ylabel("Densit di probabilit")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Gamma  link inverso
Xg = np.linspace(0.1, 2.5, 300)
beta0_g, beta1_g = 0.5, 0.3  # ensures eta > 0
eta_g = beta0_g + beta1_g * Xg
mu_g = 1 / eta_g
Yg = gamma.rvs(a=2, scale=mu_g / 2)

plt.figure(figsize=(8, 5))
plt.scatter(Xg, Yg, alpha=0.4, label="osservazioni simulate")
plt.plot(Xg, mu_g, linewidth=2, label=" = 1/(X)", color='dodgerblue')
plt.title("Gamma  link inverso (canonico)")
plt.xlabel("X")
plt.ylabel("Y positiva")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# BINOMIALE PMF
n_binom, p_binom = 10, 0.6
k_binom = np.arange(0, n_binom + 1)
plt.figure(figsize=(8, 5))
plt.bar(k_binom, binom.pmf(k_binom, n_binom, p_binom), color='dodgerblue')
plt.title("Distribuzione Binomiale (n = 10, p = 0.6)")
plt.xlabel("Numero di successi")
plt.ylabel("Probabilit")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

# Binomiale  link logit (proportion)
Xbin = np.linspace(-6, 6, 400)
beta0_bin, beta1_bin = -2, 0.8
eta_bin = beta0_bin + beta1_bin * Xbin
p_bin = 1 / (1 + np.exp(-eta_bin))
Ybin = np.random.binomial(n_binom, p_bin)
prop_success = Ybin / n_binom

plt.figure(figsize=(8, 5))
plt.scatter(Xbin, prop_success, alpha=0.4, label="prop. di successi (sim)")
plt.plot(Xbin, p_bin, linewidth=2, label="p(X) = logistic(X)", color='dodgerblue')
plt.title("Binomiale  link logit (proporzione di successi)")
plt.xlabel("X")
plt.ylabel("Proporzione di successi")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
